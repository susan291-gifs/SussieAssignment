```python
import numpy as np

class ScratchSVMClassifier:
    """
    Scratch implementation of SVM classifier
    """

    def __init__(self, num_iter, lr, kernel='linear', threshold=1e-5, verbose=False):
        self.iter = num_iter
        self.lr = lr
        self.kernel = kernel
        self.threshold = threshold
        self.verbose = verbose

    def _linear_kernel(self, Xi, Xj):
        return np.dot(Xi, Xj.T)

    def _update_lagrange_multipliers(self, lam, X, y):
        for i in range(len(lam)):
            lam_new = lam[i] + self.lr * (1 - np.sum(lam * y * y[i] * self._linear_kernel(X, X[i])))
            lam_new = max(0, lam_new)
            lam[i] = lam_new
        return lam

    def fit(self, X, y, X_val=None, y_val=None):
        """
        Learn the SVM classifier. If verification data is input, the accuracy for it is also calculated for each iteration.
        """
        self.lam_sv = np.zeros(len(X))
        for _ in range(self.iter):
            self.lam_sv = self._update_lagrange_multipliers(self.lam_sv, X, y)
            if self.verbose:
                print(f"Lagrange multipliers after iteration {_ + 1}: {self.lam_sv}")


        self.index_support_vectors = np.where(self.lam_sv > self.threshold)[0]
        self.X_sv = X[self.index_support_vectors]
        self.lam_sv = self.lam_sv[self.index_support_vectors]
        self.y_sv = y[self.index_support_vectors]
        self.n_support_vectors = len(self.index_support_vectors)
        if self.verbose:
            print(f"Number of support vectors: {self.n_support_vectors}")

        return self

    def predict(self, X):
        """
        Estimate the label using the SVM classifier.
        """

        f_x = np.zeros(len(X))


        for i in range(len(X)):
            for n in range(self.n_support_vectors):
                f_x[i] += self.lam_sv[n] * self.y_sv[n] * self._linear_kernel(X[i], self.X_sv[n])


        y_pred = np.sign(f_x)

        return y_pred
```

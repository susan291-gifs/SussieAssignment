Problem 1

```python
import math
import matplotlib.pyplot as plt

ChestNut_Diameter_cm = 2.54
ChestNut_Radius_cm = ChestNut_Diameter_cm / 2
ChestNut_Volume_cm = (4/3)* math.pi * (ChestNut_Radius_cm ** 3)
ChestNut_Volume_M = ChestNut_Volume_cm * 1e-3

Solar_System_Radius_Meter = 4.545e9
Solar_System_Volume_M = (4/3)* math.pi * (Solar_System_Radius_Meter ** 3)

volumes = []
minutes = []

def calculate_the_doubling_times(initial_volume, target_volume):
  doubling_times = 0
  while initial_volume < target_volume:
    initial_volume *=2
    volumes.append(initial_volume)
    doubling_times +=5
    minutes.append(doubling_times)
  return doubling_times

solar_time_system = calculate_the_doubling_times(ChestNut_Volume_M, Solar_System_Volume_M)
print(f"it will take {solar_time_system} minutes to cover the solar system with chestnut buns")

plt.figure(figsize=(10,6))
plt.plot(minutes, volumes)
plt.xlabel('Time (minues)')
plt.ylabel('volumes (cubic meters)')
plt.title("ChestNut Bun volume increase over time")
plt.grid(True)
plt.show()
```

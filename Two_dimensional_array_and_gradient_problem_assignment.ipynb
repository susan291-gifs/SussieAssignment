```python
import numpy as np
import matplotlib.pyplot as plt

def compute_gradient(function, x_range=(-50, 50.1, 0.1)):
    x = np.arange(x_range[0], x_range[1], x_range[2])
    y = function(x)
    gradient = np.diff(y) / np.diff(x)
    array_xy = np.column_stack((x, y))
    return array_xy, gradient

def function1(x):
    return x**2

def function2(x):
    return 2 * x**2 + 2 * x

def function3(x):
    return np.sin(np.maximum(x, 0)**0.5)

# Calculate gradients and plot graphs
for i, func in enumerate([function1, function2, function3], start=1):
    array_xy, gradient = compute_gradient(func)

    # Plot the function
    plt.figure(figsize=(12, 5))
    plt.subplot(1, 2, 1)
    plt.plot(array_xy[:, 0], array_xy[:, 1], label=f'Function {i}')
    plt.title(f'Function {i}')
    plt.xlabel('x')
    plt.ylabel('y')
    plt.legend()

    # Plot the gradient
    plt.subplot(1, 2, 2)
    plt.plot(array_xy[:-1, 0], gradient, label=f'Gradient {i}')
    plt.title(f'Gradient {i}')
    plt.xlabel('x')
    plt.ylabel('Gradient')
    plt.legend()

    plt.tight_layout()
    plt.show()
```

Problem 6

```python
import numpy as np
import matplotlib.pyplot as plt

def compute_gradient(function, x_range=(-50, 50.1, 0.1)):
    x = np.arange(x_range[0], x_range[1], x_range[2])
    y = function(x)
    gradient = np.diff(y) / np.diff(x)
    array_xy = np.column_stack((x, y))
    min_index = np.argmin(y)
    min_x, min_y = x[min_index], y[min_index]
    return array_xy, gradient, min_x, min_y, min_index

def function1(x):
    return x**2

def function2(x):
    return 2 * x**2 + 2 * x

def function3(x):
    return np.sin(np.maximum(x, 0)**0.5)

# Calculate gradients, plot graphs, and find minimum values
for i, func in enumerate([function1, function2, function3], start=1):
    array_xy, gradient, min_x, min_y, min_index = compute_gradient(func)

    # Plot the function and the gradient
    plt.figure(figsize=(15, 5))

    # Plot the function
    plt.subplot(1, 3, 1)
    plt.plot(array_xy[:, 0], array_xy[:, 1], label=f'Function {i}')
    plt.scatter(min_x, min_y, color='red', label=f'Minimum ({min_x:.2f}, {min_y:.2f})')
    plt.title(f'Function {i} and its Gradient')
    plt.xlabel('x')
    plt.ylabel('y')
    plt.legend()

    # Plot the gradient
    plt.subplot(1, 3, 2)
    plt.plot(array_xy[:-1, 0], gradient, label=f'Gradient {i}')
    plt.title(f'Function {i} Gradient')
    plt.xlabel('x')
    plt.ylabel('Gradient')
    plt.legend()

    # Display minimum value and slope before and after
    plt.subplot(1, 3, 3)
    plt.text(0.1, 0.9, f'Minimum Value: {min_y:.2f}', transform=plt.gca().transAxes)
    plt.text(0.1, 0.8, f'Slope Before Minimum: {gradient[min_index - 1]:.2f}', transform=plt.gca().transAxes)
    plt.text(0.1, 0.7, f'Slope After Minimum: {gradient[min_index]:.2f}', transform=plt.gca().transAxes)
    plt.axis('off')

    plt.tight_layout()
    plt.show()
```

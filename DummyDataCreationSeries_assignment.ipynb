Problem 1

```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(0)

mean = (-3, 0)
cova_matrix = ([[1.0, 0.8], [0.8, 1.0]])
random_num = np.random.multivariate_normal(mean, cova_matrix, size=500)

print("Generated Random Numbers:")
print(random_num[:5])
```

Problem 2

```python
plt.scatter(random_num[:, 0], random_num[:, 1], alpha=0.5)
plt.title("Random Numbers with 2D Normal Distribution")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

Problem 3

```python
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.hist(random_num[:, 0], bins=30, color='blue', alpha=0.7)
plt.title('Histogram for Dimension 1')
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.xlim(-6, 2)

plt.subplot(1, 2, 2)
plt.hist(random_num[:, 1], bins=30, color='green', alpha=0.7)
plt.title('Histogram for Dimension 2')
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.xlim(-6, 2)

plt.tight_layout()
plt.show()
```

Problem 4

```python
np.random.seed(0)

mean = (-3, 0)
cova_matrix = ([[1.0, 0.8], [0.8, 1.0]])
data_1 = np.random.multivariate_normal(mean, cova_matrix, size=500)

mean_4 = (0, -3)
cov_matrix_4 = [[1.0, 0.8], [0.8, 1.0]]

data_4 = np.random.multivariate_normal(mean_4, cov_matrix_4, size=500)

plt.scatter(data_1[:, 0], data_1[:, 1], label='Problem 1 (0)')
plt.scatter(data_4[:, 0], data_4[:, 1], label='Problem 4 (1)')

plt.title('Scatter Plot of Problems 1 and 4')
plt.xlabel('Dimension 1')
plt.ylabel('Dimension 2')
plt.legend()

plt.show()
```

Problem 5

```python
combine_data =  np.vstack((data_1, data_4))

plt.scatter(combine_data[:, 0], combine_data[:, 1], label='Combined Data')

plt.title('Scatter Plot of Combined Data')
plt.xlabel('Dimension 1')
plt.ylabel('Dimension 2')
plt.legend()

plt.show()
```

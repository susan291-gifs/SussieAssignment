### Problem 1

```python
import pandas as pd
from sklearn.datasets import load_iris
import matplotlib.pyplot as plt
import seaborn as sns
```

```python


iris = load_iris()

X = pd.DataFrame(data=iris.data, columns=iris.feature_names)

y = pd.DataFrame(data=iris.target, columns=['Species'])

print("X:")
print(X.head())

print("\ny:")
print(y.head())
```

### Problem 2

```python
df = pd.concat([X,y], axis=1)
print("df:")
print(df.head())
```

```python
print("1. 4th Sample:")
print(df.iloc[3])

print("\n2. Total Number of Samples for Each Lable:")
print(df["Species"].value_counts())

print("\n3. Missing value check:")
print(df.isnull().sum())

print("\n4. Descriptive Statistics:")
print(df.describe())
```

### Problem 6

```python

sns.set(style="whitegrid")


plt.figure(figsize=(8, 6))
label_counts = df['Species'].value_counts()
plt.pie(label_counts, labels=label_counts.index, autopct='%1.1f%%', startangle=140, colors=sns.color_palette('pastel'))
plt.title('Distribution of Samples per Label (Species)')
plt.show()


plt.figure(figsize=(12, 8))
sns.boxplot(x='Species', y='sepal_length_(cm)', data=df)
plt.title('Box Plot of Sepal Length for Each Label')
plt.show()


plt.figure(figsize=(12, 8))
sns.violinplot(x='Species', y='sepal_length_(cm)', data=df)
plt.title('Violin Plot of Sepal Length for Each Label')
plt.show()
```

### Pdroblem 7

```python
sns.set(style="ticks")
sns.pairplot(df, hue='Species', palette='pastel')
plt.suptitle('Scatter Plots of Feature Combinations (Color-coded by Species)', y=1.02)
plt.show()

sns.set(style="ticks")
sns.pairplot(df, hue='Species', palette='pastel', diag_kind='kde')
plt.suptitle('Scatterplot Matrix of Feature Combinations (Color-coded by Species)', y=1.02)
plt.show()

correlation_matrix = df[['sepal_length_(cm)', 'sepal_width_(cm)', 'petal_length_(cm)', 'petal_width_(cm)']].corr()

plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=.5)
plt.title('Heat Map of Correlation Coefficient Matrix')
plt.show()
```
